### Group By

```python
year_wise_aqi = new_df.groupby('Year')['AQI'].mean()
year_wise_aqi.plot()
```

```python
month_wise_aqi = new_df.groupby(['Year','Month'])[['AQI', 'Ozone']].mean()
flat  = month_wise_aqi.reset_index()
flat.head()
```

```python
day_wise_aqi = new_df.groupby('Date')['AQI'].agg(['mean', 'sum', 'count', 'median', 'max', 'min', 'std'])
day_wise_aqi.reset_index(inplace=True)
day_wise_aqi.head()
```

```python
unique_val = new_df.groupby('Year')['AQI'].nunique()
unique_val
```
